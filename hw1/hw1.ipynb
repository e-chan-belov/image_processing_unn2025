# 1. Загрузите изображение в оттенках серого sar_1_gray.jpg. 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.metrics import structural_similarity, mean_squared_error
image = cv2.imread("sar_1_gray.jpg", cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(5,5))
plt.imshow(image, cmap="gray")
plt.show()
```

# 2. постройте гистограмму

```python
plt.figure(figsize=(6,4))
plt.hist(image.ravel(), bins=256, range=(0,256), color="black")
plt.title("Гистограмма")
plt.xlabel("Яркость")
plt.ylabel("Кол-во пикселей")
plt.show()
```

# 3. реализуйте алгоритм гамма коррекции с параметром гамма <1, >1.

```python
def gamma_correction(img, gamma):
    corrected_pixel = np.power(img / 255.0, gamma) * 255
    return corrected_pixel.astype(np.uint8)

gamma_low = 0.5
gamma_high = 2.0

image_gamma_low = gamma_correction(image, gamma_low)
image_gamma_high = gamma_correction(image, gamma_high)

fig, axs = plt.subplots(1, 3, figsize=(15,5))
axs[0].imshow(image, cmap="gray"); axs[0].set_title("Оригинал"); axs[0].axis("off")
axs[1].imshow(image_gamma_low, cmap="gray"); axs[1].set_title(f"Gamma={gamma_low}"); axs[1].axis("off")
axs[2].imshow(image_gamma_high, cmap="gray"); axs[2].set_title(f"Gamma={gamma_high}"); axs[2].axis("off")
plt.show()
```

# 4. Сравните исходное изображение, скорректированное при помощи гамма-фильтра. MSE, SSIM.

```python
diff = []
for corrected, name in [(image_gamma_low, f"Gamma={gamma_low}"),
                        (image_gamma_high, f"Gamma={gamma_high}")]:
    mse = mean_squared_error(image, corrected)
    ssim, d = structural_similarity(image, corrected, full=True, data_range=255)
    diff.append((d * 255).astype("uint8"))
    print(f"{name}: MSE={mse:.4f}, SSIM={ssim:.4f}")
fig, axs = plt.subplots(1, 3, figsize=(15,5))
axs[0].imshow(image, cmap="gray"); axs[0].set_title("Оригинал")
axs[1].imshow(diff[0], cmap="gray"); axs[1].set_title(f"Разница при gamma={gamma_low}")
axs[2].imshow(diff[1], cmap="gray"); axs[2].set_title(f"Разница при gamma={gamma_high}")
plt.show()
```

# 5. реализуйте алгоритм статистической цветокоррекции на основе статистики eq_gray.

```python
eq_gray = cv2.equalizeHist(image)
def stat_corr_algo(src, ref): # алгоритм
    mean_src = src.mean()
    std_src = src.std()

    mean_ref = ref.mean()
    std_ref = ref.std()

    stat_corrected = (src - mean_src) * (std_ref/std_src) + mean_ref
    return np.clip(stat_corrected, 0, 255).astype(np.uint8)


image_matched = stat_corr_algo(image, eq_gray)

fig, axs = plt.subplots(1, 3, figsize=(15,5))
axs[0].imshow(image, cmap="gray"); axs[0].set_title("Оригинал")
axs[1].imshow(eq_gray, cmap="gray"); axs[1].set_title("equalizeHist")
axs[2].imshow(image_matched, cmap="gray"); axs[2].set_title("После коррекции")
plt.show()
```

# 6. Протестируйте работу алгоритмов пороговой фильтрации с различными параметрами.

```python
thresh_values = [50, 100, 150, 200]
fig, axs = plt.subplots(1, len(thresh_values), figsize=(16,5))
for i, t in enumerate(thresh_values):
    _, thresh_img = cv2.threshold(image, t, 255, cv2.THRESH_BINARY)
    axs[i].imshow(thresh_img, cmap="gray")
    axs[i].set_title(f"Threshold={t}")
plt.show()
```
